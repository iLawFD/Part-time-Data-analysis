## Clean the Date column

```python
xls_file = 'output.xlsx'
all_sheets = pd.read_excel(xls_file, sheet_name=None)  # Read all sheets into a dictionary of DataFrames

# Iterate over each sheet and modify the datetime column
for sheet_name, df in all_sheets.items():
    if 'Dates' in df.columns:  # Replace 'date_column_name' with the name of your datetime column
        df['Dates'] = df['Dates'].dt.date

# Save the modified DataFrames back to the Excel file
with pd.ExcelWriter(xls_file) as writer:
    for sheet_name, df in all_sheets.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
